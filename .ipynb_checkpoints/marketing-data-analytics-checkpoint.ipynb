````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
````

````python
def preliminary_data_cleaning(data_source):
    all_data = pd.read_csv(data_source)
    shape = all_data.shape
    missing_values = all_data.isna().sum()
    unique_values = all_data.nunique()
    all_data_cleaned = all_data.drop_duplicates()
    
    if shape[0] == len(all_data_cleaned):
        print("No duplicates were found.\n\nDO NOT FORGET TO MANUALLY CHECK FOR UNTIDY DATA!!")
    else:
        print('Duplicates found. Must manually investigate.')
    print(f'''
          \nAll DATA.head():
          \n{all_data.head(10)}\n
          \nDataframe SHAPE: {shape}\n
          \nMISSING VALUE COUNTS:
          \n{missing_values}\n
          \nUNIQUE VALUE COUNTS:
          \n{unique_values}\n
          \n\nDATA TYPES:
          \n{all_data.dtypes}
          ''')

````

sns.residplot()

Lets determine if regression analysis is appropriate for these relationships before continuing. Each plot should show: 

1. **Linearity**: 
2. **Homoscedasticity**: Variance of the dependent variable should be similar across all values of the independent variable. (spread should be the same across all income levels)
3. **Independence**: 
4. **Normal Distribution of Errors** (residuals should be normally distributed)

Based on the visuallizations created, there is some variability in purchasing behavior across different income levels. However the relationships do not necessarily show a clear linear relationship given the spread and distribution of points. 

**To Assess this** 
1. Pearson  Correlation Coefficient - understand *strength* and *direction* of relationship 
2. Simple Linear Regression - assessing the *significance* of the regression model and coefficients. 
    - will let us check the residuals for *normality* and *Homoscedasticity*


````python
from scipy.stats import pearsonr
import statsmodels.api as sm

# Initializing a dictionary to store the results
correlation_results = {}

# Looping through each purchase type column to calculate Pearson correlation coefficient
for col in num_purchases_cols:
    corr, p_value = pearsonr(marketing_data.income, marketing_data[col])
    # Add to results
    correlation_results[col] = {'correlation_coefficient': corr, 'p_value': p_value}

# Performing simple linear regression and storing the results
regression_results = {}
for col in num_purchases_cols:
    X = marketing_data['income']
    y = marketing_data[col]
    X = sm.add_constant(X) # adding a constant
    model = sm.OLS(y,X).fit()
    predictions = model.predict(X)
    regression_results[col] = model.summary()

correlation_results, regression_results['num_deals_purchases']  # Displaying results for one category as an example
````

````python
for col in num_purchases_cols:
    print(correlation_results, regression_results[col])
````

### **Simple Linear Regression `num_deals_purchases`**:



1. **Dependent Variable**: `num_deals_purchases`. This is the variable we are trying to predict or explain.

2. **R-squared**: 0.012. This value is very low, indicating that only 1.2% of the variability in the number of deals purchases is explained by income. This suggests a weak model fit.

3. **F-statistic**: 26.18 with a P-value of ~3.37e-07. This indicates that* the model is statistically significant*, but given the low R-squared, it's *not very useful for prediction*.

4. **Coefficients**: 
  - **Constant (Intercept)**: 2.8272. This is the expected number of deals purchases when income is zero.
  - **Income Coefficient**: -9.674e-06. This suggests that for each unit increase in income, the number of deals purchases decreases slightly. However, the effect is very small.

5. **Diagnostic Statistics**: 
  - **Omnibus, Skew, Kurtosis, and Jarque-Bera (JB)**: These tests suggest that the *residuals are not normally distributed*, as indicated by the significant Omnibus and JB tests and the high skew and kurtosis.
  - **Durbin-Watson**: 1.757, which is close to 2, suggests *no major concern regarding autocorrelation*.

- **Notes**: The large condition number may indicate multicollinearity or numerical problems. This can be a concern in regression analysis and might need further investigation.

#### Conclusion
While there is some statistically significant correlation between income and various types of purchases, the linear regression models, especially for 'num_deals_purchases', are weak in explaining a significant portion of the variability. The data might not be best suited for linear regression modeling, or it might need more complex models or different variables for a better fit.


Let's interpret the linear regression results for `num_web_purchases`:

#### OLS Regression Results `num_web_purchases`

1. **Dependent Variable**: `num_web_purchases`. This is the variable we are trying to predict or explain.

2. **R-squared**: 0.211. This value suggests that about 21.1% of the variability in the number of web purchases is explained by the model. This is a moderate level of explanation, significantly higher than what we saw with `num_deals_purchases`.

3. **F-statistic**: 590.7 with a P-value of ~7.64e-116. This indicates that the *model is statistically significant*. The extremely low P-value suggests that *income is a significant predictor of the number of web purchases.*

4. **Coefficients**: 
   - **Constant (Intercept)**: 1.0485. This represents the expected number of web purchases when income is zero.
   - **Income Coefficient**: 5.85e-05. This indicates a positive relationship between income and the number of web purchases. For each unit increase in income, the number of web purchases is expected to increase by this coefficient.

5. **Model Fit and Diagnostic Statistics**:
   - **Adjusted R-squared**: 0.211, very close to the R-squared, which is good as it indicates the model is not unduly influenced by the number of predictors.
   - **Log-Likelihood, AIC, and BIC**: These values are used for model comparison and complexity assessment. Lower values generally indicate a better model, but they are more useful when comparing different models.
   - **Omnibus, Skew, Kurtosis, and Jarque-Bera (JB)**: The Omnibus and JB tests are significant, suggesting that the *residuals are not normally distributed*. The *high Skew and Kurtosis* values also point towards this. 
   - **Durbin-Watson**: 1.845, which suggests minimal concern regarding autocorrelation in the residuals.
   - **Cond. No.**: 1.47e+05. A large condition number, indicating potential multicollinearity or other numerical problems, similar to the previous model.

#### Conclusion
The model for `num_web_purchases` shows a moderate explanatory power with a significant positive relationship between income and the number of web purchases. However, like the previous model, the large condition number and the distribution of residuals indicate that caution should be exercised in interpretation, and possibly more complex modeling techniques might be required for a better fit.

    Clean your data. Make sure that all the variables and values make sense.
    Decide how you will handle missing data. 
    Concatenate or split values as necessary to work with your variables. 
    Group and merge as needed to get data at the aggregation level you will need.

Once your data is cleaned and ready to go, explore it through the use of *summary statistics* and *data visualizations*. Make sure to document findings along the way.

- **Find the story**
What, in one sentence, do you want your audience to take away from your project? You may have to create all of your visualizations, do all of your explorations, and even get feedback before you know what it is. But, take a moment to decide the one thing people should take away from your project and make that the first thing. Make that story the easiest one to find.

    - Often you will start putting together your report or your dashboard and find that the story you really want to tell is the last thing in your project. Flip the narrative and lead with that. Yes, it will require remaking or making a new version, but most of the time, you will end up with a much better project.
/

- **Communicate your findings**
    This could be in the shape of a report, a Tableau Dashboard, a slide deck, etc. The best way to decide is to imagine your audience. Who are they? Where do you envision reaching them? What is the best medium for that?

    Do you want to reach other people interested in your topic? Do you want to persuade someone to care about something? Do you want to focus on business stakeholders? Where might you find each group? Tailor your project to that imagined scenario.

    From there, you can start to create a project to communicate your story with them.

- **Wrap up**
    This step might be the most impactful for finding a job. Now that youve written your report, created your deck, or built your dashboard, check your work. Go through and proofread. Try to view your project from wherever you decided to host it. Make sure that all of the links work, and ask a friend (or visit the forums) to find someone to go through it and tell you if there are any inconsistencies, jumps of logic, or confusing parts. This final round of checks is essential for creating a polished project.

- **Next steps**
    Portfolio in hand, you are ready to apply to jobs! Be sure to review all of your projects before any interviews - you never know when someone will ask you about one!
